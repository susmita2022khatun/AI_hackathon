```python
import numpy as np
#extracting metadata of training image
list_train = !exiftool  /content/gdrive/MyDrive/reva/data/images/train -gpsaltitude -gpslatitude -gpslongitude
list_train = np.array(list_train)
#extracting metadata of val image
list_val = !exiftool  /content/gdrive/MyDrive/reva/data/images/val -gpsaltitude -gpslatitude -gpslongitude
list_val = np.array(list_val)
#extracting metadata of test image
list_test = !exiftool  /content/gdrive/MyDrive/reva/data/images/test -gpsaltitude -gpslatitude -gpslongitude
list_test = np.array(list_test)
#extracting metadata of predict image
list_predict = !exiftool  /content/gdrive/MyDrive/reva/data/images/predict -gpsaltitude -gpslatitude -gpslongitude
list_predict = np.array(list_predict)
```

```python
import os
#function for extracting location of val data
def location_val(image_name):
  path = os.path.join("======== /content/gdrive/MyDrive/reva/data/images/val", image_name)
  for i in range(len(list_val)):
    if path== list_val[i]:
      location_val = np.array([list_val[i+1],list_val[i+2],list_val[i+3]])
  return location_val

#function for extracting location of test data
def location_test(image_name):
  path = os.path.join("======== /content/gdrive/MyDrive/reva/data/images/test", image_name)
  for i in range(len(list_test)):
    if path== list_test[i]:
      location_test = np.array([list_test[i+1],list_test[i+2],list_test[i+3]])
  return location_test

##function for extracting location of train data
def location_train(image_name):
  path = os.path.join("======== /content/gdrive/MyDrive/reva/data/images/train", image_name)
  for i in range(len(list_train)):
    if path== list_train[i]:
      location_train = np.array([list_train[i+1],list_train[i+2],list_train[i+3]])
  return location_train

#function for extracting location of predict data
def location_predict(image_name):
  path = os.path.join("======== /content/gdrive/MyDrive/reva/data/images/predict", image_name)
  for i in range(len(list_predict)):
    if path== list_predict[i]:
      location_predict = np.array([list_predict[i+1],list_predict[i+2],list_predict[i+3]])
  return location_predict
```

```python
import os
#function for counting number of boxes
def num_of_object(image_name):
  result_out = model.predict(image_name)
  result_out2 = result_out[0]
  k = len(result_out2.boxes)

  return k
```
